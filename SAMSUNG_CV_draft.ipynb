```python
kernel = torch.tensor([[[[0., 1, 0],
                         [1,  2, 1],
                         [0,  1, 0]],

                        [[1, 2, 1],
                         [0, 3, 3],
                         [0, 1, 10]],

                        [[10, 11, 12],
                         [13, 14, 15],
                         [16, 17, 18]]]])
```

```python
def calc_out_shape(input_matrix_shape, out_channels, kernel_size, stride, padding):
    batch_size, channels_count, input_height, input_width = input_matrix_shape
    output_height = (input_height + 2 * padding - (kernel_size - 1) - 1) // stride + 1
    output_width = (input_width + 2 * padding - (kernel_size - 1) - 1) // stride + 1

    return batch_size, out_channels, output_height, output_width  
```

```python
out_shape = calc_out_shape(x.shape, kernel.shape[0], kernel.shape[1], 1, 0)
out_shape
```

```python
y = torch.zeros(out_shape)
y
```

```python
for i in range(out_shape[0]):
  for j in range(out_shape[1]):
    for k in range(out_shape[2]):
      for l in range(out_shape[3]):
        print((x[i, :, k * 1 : k * 1 + kernel.shape[2], l * 1 : l * 1 + kernel.shape[3]] * kernel).sum())
        y[i, j, k, l] = (x[i, :, k * 1 : k * 1 + kernel.shape[2], l * 1 : l * 1 + kernel.shape[3]] * kernel).sum()
```
